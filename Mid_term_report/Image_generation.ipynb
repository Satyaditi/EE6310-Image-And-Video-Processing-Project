```python
##importing required libraries
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw, ImageFilter
import numpy as np

from scipy.ndimage import rotate, shift
from scipy.signal import convolve2d

import random

import pywt
from scipy.interpolate import interp2d

import cv2
```

```python
#image filepath
#file_path = "/content/hydra_poss.gif"
file_path = "D:/B.TECH/CLASSES/3rd Year/SEM 6/IVP/Project/2/hydra_poss.gif"

I_clean = Image.open(file_path).convert('L')

fig = plt.subplots(1,1,figsize=(8,8))
plt.imshow(I_clean,cmap="gray")

plt.show()
```

```python
def pixel_reduction(I, size):
    img = Image.fromarray(np.uint8(I))
    img = img.resize((size, size), resample=Image.LANCZOS)
    return img
```

```python
I_clean = pixel_reduction(I_clean,256)

fig = plt.subplots(1,1,figsize=(8,8))
plt.imshow(I_clean,cmap="gray")

plt.show()
```

```python
def adding_trail(I_clean, x1,y1,x2,y2):
    draw = ImageDraw.Draw(I_clean)

    line_color = 255
    draw.line((x1, y1, x2, y2), fill=line_color, width=2)

    I_trail = np.array(I_clean, dtype=np.float32)
    
    return I_trail
```

```python
I_trail = adding_trail(I_clean, x1=20, y1=0, x2=160, y2=383)

fig = plt.subplots(1,1,figsize=(8,8))
plt.imshow(I_trail,cmap="gray")

plt.show()
```
